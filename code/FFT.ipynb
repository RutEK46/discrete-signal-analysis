```python
from numpy import *
import plotly as py
py.offline.init_notebook_mode(connected=True)


frequencyList = [1, 30]


def dft(X):
    Y = []
    for k in range(len(X)):
        y = 0
        for n in range(len(X)):
            ex = -2j * pi * k * n / len(X)
            y += X[n] * exp(ex)
        Y.append(y)
    return Y


def fft(X):
    if len(X) == 1:
        return X
    else:
        X = fft([X[i] for i in range(0, len(X), 2)]) + fft([X[i] for i in range(1, len(X), 2)])
        Y1 = []
        Y2 = []
        for k in range(int(len(X) / 2)):
            t = X[k]
            n = len(X)
            ex = -2j * pi * k / n
            Y1.append(t + exp(ex) * X[k + int(n/2)])
            Y2.append(t - exp(ex) * X[k + int(n/2)])
        return Y1 + Y2


X = arange(0, 128, 0.25)

finalSignal = [sum([sin(frequency * x) for frequency in frequencyList]) for x in X]
dtfResult = dft(finalSignal)
fftResult = fft(finalSignal)

trace1 = dict(
        visible = True,
        line = dict(color='#F000FF', width=1),
        name = 'Fourier transform',
        x = X,
        y = [abs(number) for number in dtfResult])

trace2 = dict(
        visible = True,
        line = dict(color='#00F0FF', width=1),
        name = 'Fast Fourier transform',
        x = X,
        y = [abs(number) for number in fftResult])

fig = dict(data=[trace1, trace2])

py.offline.iplot({
    "data": fig
})
```
