```python
CSV_FILE_NAME = 'wig20_d.csv'
CSV_KEY = 'Otwarcie'
```

```python
import plotly as py
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import itertools as it

py.offline.init_notebook_mode(connected=True)

data0 = pd.read_csv(CSV_FILE_NAME)[CSV_KEY]
x = []

for i in range(len(data0)):
    x.append(i)

trace0 = dict(
        visible = False,
        line=dict(color='#0000FF', width=1),
        name = 'data',
        x = x,
        y = data0)

trace1 = (dict(
        visible = True,
        line=dict(color='#FFA500', width=1),
        name = 'macd',
        x = x,
        y = data0.ewm(span=12, adjust=False).mean() - data0.ewm(span=26, adjust=False).mean()))

trace2 = (dict(
        visible = True,
        line=dict(color='#FF0000', width=1),
        name = 'signal',
        x = x,
        y = trace1['y'].ewm(span=9, adjust=False).mean()))

fig = dict(data=[trace1, trace2])

py.offline.iplot({
    "data": fig})
```
